```python
import sqlite3
db_file = 'Data Warehousing.db'
try:
    conn = sqlite3.connect(db_file)
    print(f"Successfully created/connected to database: {db_file}")

except sqlite3.Error as e:
    print(f"An error occurred: {e}")
finally:
    if conn:
        conn.close()
        print("Database connection closed.")
```
